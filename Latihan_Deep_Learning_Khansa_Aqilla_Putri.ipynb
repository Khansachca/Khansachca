Features/variabel pada data diabetes adalah 8
Memiliki 2 class
Input layer: 8 neuron
Output layer: 2 neuron

```python
import pandas as pd
data = pd.read_csv('/content/diabetes_dataset.csv')
print(data)
x = data.values[:, :-1]
y = data.values[:, -1]
x[0:8]
x = x.astype('int')
x[0:8]

from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=30, random_state=42)

print(' Ukuran x train: ', x_train.shape)
print(' Ukuran y train: ', y_train.shape)
print()
print(' Ukuran x test: ', x_test.shape)
print(' Ukuran y test: ', y_test.shape)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import Sequential, Model, Input
from tensorflow.keras.layers import Dense

model = Sequential(name='Sequential_API_2')
model.add(Input(shape=(8,)))
model.add(Dense(8, activation='tanh'))
model.add(Dense(3, activation='tanh'))
model.add(Dense(1, activation='softmax'))

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
model.compile(
    optimizer = Adam(learning_rate=0.001),
    loss = BinaryCrossentropy(),
    metrics = ['Accuracy']
)
hist = model.fit(
    x = x_train,
    y = y_train,
    validation_data = (x_test, y_test),
    batch_size = 32,
    epochs = 130,
    verbose = 2
)
```
